```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
%matplotlib inline
```

```python
# np.linspace() - > Return evenly spaced numbers over a specified interval.
data=np.linspace(-10,10,100)
```

### sigmoid activation function

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
## Sigmoid plot
sns.lineplot(x=data,y=sigmoid(data),color='b',label="sigmoid")
```

### Tanh activation funtion

```python
def tanh(z):
    return np.tanh(z)
```

```python
# tanh plot
sns.lineplot(x=data,y=tanh(data),color='r',label='tanh')
```

### ReLu activation function (Rectified Linear Unit)

```python
def reLu(z):
    return np.maximum(0,z)
```

```python
## reLU plot

sns.lineplot(x=data,y=reLu(data),color='y',label='reLu')
```

### Leaky reLu

```python
def Leaky_reLu(z,alpha=0.01):
    return np.maximum(alpha*z,z)
```

```python
sns.lineplot(x=data,y=Leaky_reLu(data),color='#3DBE29',label='Leaky reLu')
```

### ELU activation function

```python
# alpha is like a hyper parameter

def elu(z,alpha=1):
    return np.where(z>0 ,z,alpha*(np.exp(z)-1))
```

```python
sns.lineplot(x=data,y=elu(data),color='#FF6666',label='Leaky reLu')
```

#### Plotting all activation funtions in one plot

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.lineplot(x=data,y=sigmoid(data),color='b',label="sigmoid")
sns.lineplot(x=data,y=tanh(data),color='r',label='tanh')
sns.lineplot(x=data,y=reLu(data),color='y',label='reLu')
sns.lineplot(x=data,y=Leaky_reLu(data),color='#3DBE29',label='Leaky reLu')
sns.lineplot(x=data,y=elu(data),color='#46B2E0',label='Leaky reLu')
```

<h1><font color='red'>Advantages and disadvantages</font></h1>

- **Sigmoid Activation function**
    - **Advantages**<br>
        - Sigmoid function gives advantage of clear prediction(0 or 1) in binary classifiers.
    - **Disadantages**<br>
        - Sigmoid function prones to *vanishing gradient descent probelm*.
        - And the function output is not zero centric.

- **Tanh Activation function**
    - **Advantages**
        - Function output is Zero centric.
    - **Disadvantages**
        - Time complexity is more in calculation the derivatives in Backward propagation.
        - *Vanishing gradient descent problem* still exists in Deep layered networks
        
- **ReLu Activation funtion**<br>
    - **Advantages**<br>
        - Function outputs results in Linear relationship.
        - Solves vanishing gradient descent problem
        - Faster than Sigmoid and Tanh activation functions
    - **Disadvantages**
        - If the derivative is equal to 0 then it cause *Dead Neurons*
        - Function output is not zero centric.
        
- **Leaky ReLu Activation function**
    - **Advantages**
        - Prevents the dead neuron problem. By introducing the *alpha(small value)* as Hyper parameter.
    - **Disadvantage**
        - Function output is not zero centric.
        
- **ELU (Exponential Linear Units)**
    - **Advantages**
        - Solves *vanishing Gradient descent* probelm.
        - Function output is zero centric.
    - **Disadvantages**
        - Time complexity is more
        
        

<h2><font color='#23C4ED'>Vanishing Gradient Descent problem</font></h2>


   - When we are updating the weights to reduce the loss function in backward propogration. Vanishing gradient descent problem leads to that the Old and Updated new weights are approximately same. And the convergence will never or takes more time to reach the global minima.
   
   - *Example* 
      - The derivative of sigmoid function output always ranges between **0 - 0.25**. So this leads to the vanishing gradient descent problem in backward propogation when using sigmoid function
